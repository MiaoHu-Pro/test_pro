## 划分训练集，验证集，测试集

```python
def split_train_test_valid():
    # read file
    # input_path = "E:\\Data\\"
    # file = "relationship.csv"
    df_flow = pd.read_csv("relationship.csv", header=None, encoding='gbk')

    # define the ratios :train:5/8, valid1/8, test:2/8
    train_len = int(len(df_flow) * 0.625)
    test_len = int(len(df_flow) * 0.25)

    # split the dataframe
    idx = list(df_flow.index)
    random.shuffle(idx)  # 将index列表打乱
    df_train = df_flow.loc[idx[:train_len]]
    df_test = df_flow.loc[idx[train_len:train_len+test_len]]
    df_valid = df_flow.loc[idx[train_len+test_len:]]  # 剩下的就是valid

    # output
    df_train.to_csv('train.txt', header=False, index=False, sep='\t')
    df_test.to_csv('test.txt', header=False, index=False, sep='\t')
    df_valid.to_csv('valid.txt', header=False, index=False, sep='\t')
```

```python
split_train_test_valid()
```
